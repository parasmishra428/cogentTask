```python
def status_locaton(st1,loc):
  if(loc == 'A'):
    if(st1 == 'dirty'):
      print("location: ",loc," status: ",st1)
      print("Cleaning location A....")
      print("move left to location: B")
    else:
      print("location: ",loc," status: ",st1)
      print(" A location is already cleaned")
      print("move left to location: B")

  if(loc == 'B'):
    if(st1 == 'dirty'):
      print("Location: ",loc," status: ",st1)
      print("Cleaning location B....")
      print("Move Right to location: A")
    else:
      print("location: ",loc," status: ",st1)
      print(" B location is already cleaned")
      print("Move Right to location: A")
```

```python
import random

status = random.choice(['dirty','clean'])
location = random.choice(['A','B'])
print("location: ",location)
print("status: ",status)
```

Write a program to illustrate the DFS

<b> The DSF algorithm follows as:</b>

1. We will start by putting any one of the graph's vertex on top of the stack.
2. After that take the top item of the stack and add it to the visited list of the vertex.
3. Next, create a list of that adjacent node of the vertex. Add the ones which aren't in the visited list of vertexes to the top of the stack.
4. Lastly, keep repeating steps 2 and 3 until the stack is empty.



image.png


```python
graph = {
  '5' : ['3','7'],
  '3' : ['2', '4'],
  '7' : ['8'],
  '2' : [],
  '4' : ['8'],
  '8' : []
}

visited = set() # Set to keep track of visited nodes of graph.

def dfs(visited, graph, node):  #function for dfs
    if node not in visited:
        print (node)
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)

# Driver Code
print("Following is the Depth-First Search")
dfs(visited, graph, '5')
```

<b>Write a program to illustrate BFS</b>


<br>
<br>

<b>The steps of the algorithm work as follow:</b>

 1. Start by putting any one of the graphs vertices at the back of the queue.
 2. Now take the front item of the queue and add it to the visited list.
 3. Create a list of that vertex's adjacent nodes. Add those which are not within the visited list to the rear of the queue.
 4. Keep continuing steps two and three till the queue is empty.



```python
graph = {
  '5' : ['3','7'],
  '3' : ['2', '4'],
  '7' : ['8'],
  '2' : [],
  '4' : ['8'],
  '8' : []
}

visited = [] # List for visited nodes.
queue = []     #Initialize a queue

def bfs(visited, graph, node): #function for BFS
  visited.append(node)
  queue.append(node)
  print(" node:",node)

  while queue:          # Creating loop to visit each node
    m = queue.pop(0) # extract the node from queue using pop
    print (m)
    print("queue" ,queue)

    for neighbour in graph[m]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)

# Driver Code
print("Following is the Breadth-First Search")
bfs(visited, graph, '5')    # function calling
```

## <b> write a basic program of neuron network using dot product and vector addition </b>

image.png

```python
# single neuron:
import numpy as np
inputs = [ 1.0 , 2.0 , 3.0 , 2.5 ]
weights = [ 0.2 , 0.8 , - 0.5 , 1.0 ]
bias = 2.0
outputs = np.dot(weights, inputs) + bias
print("Output of the neuron is")
print (outputs)
```

```python
#multi layer of neuron
import numpy as np
inputs = [ 1.0 , 2.0 , 3.0 , 2.5 ]
weights = [
[ 0.2 , 0.8 , - 0.5 , 1 ],
[ 0.5 , - 0.91 , 0.26 , - 0.5 ],
[ - 0.26 , - 0.27 , 0.17 , 0.87 ]
]
biases = [ 2.0 , 3.0 , 0.5 ]
layer_outputs = np.dot(weights, inputs) + biases
print("Output of the neurons n1,n2 & n3 are:")
print (layer_outputs)
```
